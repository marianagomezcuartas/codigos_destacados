```python
#definimos algunas funciones para convertir a unidades mas apropiadas

def conversion_horas(dat):                                               #donde dat=(horas, minutos, segundos)
  seg_a_min = float(dat[2]) / 60                                         #un minuto son 60seg
  min = float(dat[1]) + seg_a_min
  horas = float(dat[0]) + (min / 60)                                        #una hora son 60min
  return horas

def tiempo_a_grados(dat):
  grados= float(dat) *15
  return grados

def conversion_grados(dat):                                         #donde dat= (signo, grados, minutos, segundos)
  seg_a_min = float(dat[3]) / 60                                           #un minuto son 60''
  min = float(dat[2]) + seg_a_min
  grados = float(dat[1]) + (min / 60)                                      #un grados son 60'
  return grados * dat[0]

def separacion(dato):                                              #además, establezco una función tal que pueda separar por comas las horas, minutos y segundos para la ascension recta
  horas= dato[0:3]
  minutos= dato[3:6]
  segundos= dato[6:]
  return (horas, minutos, segundos)

def separacion_dec(dato):
  grados= dato[1:4]
  minutos= dato[4:7]
  segundos= dato[7:]
  if dato[0] == '-':
    signo = -1
  elif dato[0] == '+':
    signo = 1
  return signo, grados, minutos, segundos

def msa_to_grados(dato):
  seg_arc = dato * (1*10**-3)
  grados = seg_arc /3600
  return grados

def msa_to_sa(dato):
  seg_arc= dato * (1*10**-3)
  return seg_arc
```

```python
#comienzo a trabajar sobre el dataframe

df['AR(°)'] = df['AR(°)'].apply(separacion)
df['DEC(°)'] = df['DEC(°)'].apply(separacion_dec)
df['AR(°)']= df['AR(°)'].apply(conversion_horas)
df['AR(°)'] = df['AR(°)'].apply(tiempo_a_grados)
df['DEC(°)'] = df['DEC(°)'].apply(conversion_grados)
df['mu_ar*cos(dec)(°/yr)'] = df['mu_ar*cos(dec)(°/yr)'].apply(msa_to_grados)
df['mu_dec(°/yr)'] = df['mu_dec(°/yr)'].apply(msa_to_grados)
df['paralaje(")']  =df['paralaje(")'].apply(msa_to_sa)
df
```

```python
#Defino las funciones de correcciones y algunas otras que son útiles en medio del código

import numpy as np
rad = np.pi / 180

def dia_juliano(fecha):
  '''Ingrese la fecha de la forma "yy-mm-dd"'''
  fecha= pd.to_datetime(fecha)
  Y = fecha.year
  M = fecha.month
  D= fecha.day
  if M>2:
    Y = Y
    M = M
  elif M==1 or M==2:
    Y= Y-1
    M= M + 12
  A= int(Y/100)
  B= 2 - A + int(A/4)
  JD= int(365.25 *(Y + 4716)) + int(30.6001 *(M + 1)) + D + B - 1524.5
  return JD                                                                     #me regresa el valor a las cero horas

def corr_precesion(fecha, dec0, ar0):
  '''ingrese la fecha en formato "yy-mm-dd" , y lo demás en grados'''
  fecha= pd.to_datetime(fecha)
  FJ = dia_juliano(fecha)
  J2000 = 2451545
  T = (FJ - J2000)/(36525)

  M = 1.2812323 * T + 3.879 * (10**-4) * T**2 + 1.101 * (10**-5) * T**3         ##en grados
  N = 0.5567530 * T - 1.185 * (10**-4) * T**2 - 1.16 * (10**-5) * T**3

  arm = ar0 + (1/2) * (M + N * (np.sin(ar0 * rad) * np.tan(dec0 * rad)))
  decm = dec0 + (1/2) * N * np.cos(arm * rad)

  ar = ar0 + M + (N * np.sin(arm*rad) * np.tan(decm * rad))
  dec= dec0 + (N * np.cos(arm * rad))

  return  dec , ar                 ##regresa en grados


def primera_circunferencia(grados):
  if grados > 360:
    grados = grados - int(grados/360)*360
  if grados < 0:
    grados = 360 + grados - int(grados/360)*360
  return grados


def corr_nutacion(fecha, dec, ar):
  '''ingresar los valores en grados'''
  FJ = dia_juliano(fecha)
  J2000 = 2451545
  T = (FJ - J2000)/(36525)

  om= primera_circunferencia(125.04 - 1934.13 * T )            #grados
  D = primera_circunferencia(297.85 + 445267.11 * T)
  F = primera_circunferencia(93.27 + 483202.0175 * T)

  del_long = -17.2/3600 * np.sin(om*rad) + 0.2/3600 * np.sin( 2 * om * rad) - 1.3/3600 * np.sin((2*om + 2*F - 2*D)*rad) - 0.2/3600 * np.sin((2*om + 2*F)*rad)
  del_eps = 9.2/3600 * np.cos(om*rad) - 0.1/3600 * np.cos(om*rad) + 0.6/3600 * np.cos((2*om + 2*F - 2*D)*rad) + 0.1/3600 * np.cos((2*om + 2*F)*rad) #grados

  eps= 23 + 26/60 + 21.4/3600 - (46.81/3600) * T

  del_ar = (np.cos(eps * rad) + np.sin(eps * rad) * np.sin(ar * rad) * np.tan(dec * rad)) * del_long - np.cos(ar * rad) * np.tan(dec * rad) * del_eps
  del_dec = np.sin(eps * rad) * np.cos(ar * rad) * del_long + np.sin(ar * rad) * del_eps

  ar_v= ar + del_ar
  dec_v = dec + del_dec
  eps_v = eps + del_eps              #puedo obtenerlo de la función en caso de ser necesario

  return dec_v , ar_v        #todo sale en grados

def corr_mov_propio(fecha, en_dec, en_ar, dec, ar):
  FJ = dia_juliano(fecha)
  J2000 = 2451545
  years = (FJ - J2000)/(365.25)
  cambio_ar = en_ar * years
  cambio_dec = en_dec * years
  ar_corr = ar + cambio_ar
  dec_corr = dec + cambio_dec
  return dec_corr, ar_corr
```

```python
#Veamos ahora una funcion que resuma las tres correciones a un solo resultado

def prec_nut_mov(fecha, dec0, ar0, mu_dec, mu_ar):
  '''ingrese la fecha en formato "yy-mm-dd" , dec y ar en grados, y los años transcurridos desde el año J2000'''
  fecha= pd.to_datetime(fecha)
  c_prec = corr_precesion(fecha, dec0, ar0)
  c_nut = corr_nutacion(fecha, c_prec[0], c_prec[1])
  c_mov = corr_mov_propio(fecha, mu_dec, mu_ar, c_nut[0], c_nut[1])
  return c_mov[0], c_mov[1]        #dec, ar
```

```python
#ahora bien, lo que resta es encontrar los cambio en ascension recta y declinacion, considerando cada una de las correcciones mencionadas

def cambios(corr_dec, corr_ar, dec0, ar0):
  del_dec = corr_dec - dec0
  del_ar = corr_ar - ar0
  return del_dec, del_ar
```

Para comprender porqué esto sucede, podemos realizar una analogía con eventos de la cotidianidad. Por ejemplo, si vemos un objeto que se encuentra a una distancia 'corta' moverse con una velocidad v y otro objeto del mismo tamaño ubicado a una distancia mucho más alejada que el primero, y que se mueve a su misma velocidad, entonces va ser más complicado percibir el movimiento del segundo en comparación con el primero debido mayormente a la distancia que los separa del punto de observación.\\
Para el caso de las estrellas sucede algo similar, donde a mayor distancia que se ubique una estrella, más complicado es observar e identificar su movimiento propio.

```python
#Vamos ahora a calcular la velocidad tangencial de las estrellas

def velocidad_tangencial(dist, mov_propio):
  '''Ingresar la distancia en parsecs, y el movimiento propio en °/año'''      #paso a radianes, para que al final no me queden involucradas unidades algulares

  #comencemos por convertir mov_propio de °/año, a rad/año
  mov_propio = mov_propio * np.pi / 180

  dist_recorrida = dist * np.sin(mov_propio)                                                   #distancia que la estrella recorre en un año
  dist = dist_recorrida

  return dist         #velocidad tangencial de la estrella en pc/año
```

```python
AU = 149597870.7    #en km
def km_s_to_pc_año(dat):
  result = (dat * 3600 * 24 * 365.25) / (AU * 206265)
  return result
```

```python
vel_tan['velocidad radial (pc/año)'] = vel_tan['velocidad radial (km/s)'].apply(km_s_to_pc_año)
```

```python
def cambio_mov_propio(years, vel_rad, dist, v_tan):
  dist_cambio = years * vel_rad
  dist_tot = dist + dist_cambio
  cambio_mov_propio = v_tan / dist_tot
  return cambio_mov_propio * 180/np.pi
```

###altura max. geometrica y aparente

```python
#Ahora, consideremos un código con el cual podamos determinar la altura máxima geométrica y aparente de un astro,
#suponiendo que tenemos condiciones sobre el nivel del mar


def corr_refraccion(h_app , presion, temp):
  '''ingrese T en °cent, P en atm, Ha en grados'''
  Re = ((0.28 * presion * 1013)/(temp + 273)) * (0.0167/np.tan(rad * (h_app + (7.31/ (h_app + 4.4)))))     ##donde 1atm = 1013 milibares

  h_verdadera = h_app - Re
  return h_verdadera


def alt_ap_geo(dec, lat):          #considero condiciones sobre el nivel del mar: presion= 1atm, temperatura=25°c
  presion=1
  temperatura =25
  value= lat - dec
  if value>0:
    h_max_app = 90- value
  else:
    h_max_app = 90+ value

  h_geometrica = corr_refraccion(h_max_app, presion, temperatura)

  return h_max_app, h_geometrica
```

###hora de salida, puesta y culminación

```python
#veamos primero la hora de culminación.
#luego determinamos si las estrellas son circumpolares, anticircumpolares o ninguna de las anteriores, para hallar su hora de salida y puesta

def hora_culminacion(fecha, ar, lon ):
  '''ingresar ascensión recta y longitud en grados'''
  HH = lon / 15
  ar = ar/ 15      #paso a horas
  f = 1.0027379

  TSL = ar
  TSGo = TSG0(fecha)
  TSGt = primer_reloj(ar - (lon / 15))
  TU = primer_reloj((TSGt - TSGo) / f)
  TL = primer_reloj(TU + HH)

  return TL
```

```python
def hora_sp(fecha,lat,lon, dec, ar):
  '''ingresar cada uno de los valores en grados'''
  #veamos primero si es circumpolar o anticircumpolar
  if lat>0:
    if dec > 90- lat:
      circumpolar = True
      anticircumpolar = False
    elif dec < lat - 90:
      circumpolar = False
      anticircumpolar = True
    else:
      circumpolar = False
      anticircumpolar = False
  elif lat<0:
    if dec < -90-lat:
      circumpolar = True
      anticircumpolar =False
    elif dec > 90 + lat:
      circumpolar = False
      anticircumpolar = True
    else:
      circumpolar = False
      anticircumpolar = False

  if circumpolar == True or anticircumpolar ==True:
    hora_s = 'No aplica'
    hora_p = 'No aplica'
    if circumpolar == True:
      detalle = 'No se oculta'
    elif anticircumpolar ==True:
      detalle = 'No sale'
  if circumpolar == False and anticircumpolar ==False:
    detalle = 'Si sale'
    HH = lon / 15
    ar = ar/ 15      #paso a horas
    f = 1.0027379

    hs = (-34) / 60           #en este caso no vamos a corregir por diámetro aparente de las estrellas, debido a la distancia a la que se encuentran
    Hp = np.arccos((np.sin(hs * rad)- np.sin(dec * rad) * np.sin(lat * rad))/(np.cos(dec * rad) * np.cos(lat * rad))) * (1/rad)
    Hs= 360 - Hp   ###sale en gradossss --> abajo cambiar a horas

    TSGo = TSG0(fecha)

    TSLs= primer_reloj(ar + Hs/15)
    TSGts = primer_reloj(TSLs - (lon/15))
    TUs = primer_reloj((TSGts - TSGo) / f)
    hora_s = primer_reloj(TUs + HH)

    TSLp= primer_reloj(ar + Hp/15)        #####ojo en horas los dossss
    TSGtp = primer_reloj(TSLp - (lon/15))
    TUp = primer_reloj((TSGtp - TSGo) / f)
    hora_p = primer_reloj(TUp + HH)

  return hora_s , hora_p , detalle
```

```python
def culm_s_p(fecha, lat, lon, dec, ar):
  hora_culminacion = hora_culminacion(fecha, ar, lon)
  hora_salida = hora_sp(fecha,lat,lon, dec, ar)[0]
  hora_puesta = hora_sp(fecha,lat,lon, dec, ar)[1]
```

###Duración del día

```python
def duracion_dia(hora_s, hora_p, detalle):
  if detalle == 'No se oculta':
    duracion_dia = 24
  if detalle =='No sale':
    duracion_dia = 0
  if detalle =='Si sale':
    duracion_dia = hora_p - hora_s

  return duracion_dia
```

###Código unificador

```python
def info_sol(fecha, lat, lon, dec, ar):
  angulo_hor_salida = salida_sol(fecha, lat, lon, dec, ar)[0]
  azimut_salida = salida_sol(fecha, lat, lon, dec, ar)[1]

  h_max_app = alt_ap_geo(dec, lat)[0]
  h_max_geom = alt_ap_geo(dec, lat)[1]

  hora_salida = hora_sp(fecha,lat,lon, dec, ar)[0]
  hora_puesta = hora_sp(fecha,lat,lon, dec, ar)[1]
  detalle = hora_sp(fecha,lat,lon, dec, ar)[2]       ##no imprimir este
  hora_culm = hora_culminacion(fecha, ar, lon)
  dur_dia = duracion_dia(hora_salida, hora_puesta, detalle)

  return angulo_hor_salida, azimut_salida, h_max_app, h_max_geom, hora_salida, hora_puesta, hora_culm, dur_dia
```
