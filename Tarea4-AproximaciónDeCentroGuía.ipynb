La idea ahora es escribir una rutina que nos permita calcular la posición del planeta usando la ACG.

Para ello, vamos a hacer uso de las ecuaciones 14 y 15 encontradas en la demostración del primer punto:

14. $$x\approx -ae\cos(M)$$

15. $$y\approx 2ae\sin(M)$$

````python
def posicion_ACG(M, a, e):
  '''Recibe un arreglo de valores de anomalía media (M) en radianes, semieje mayor en unidades
  canónicas y excentricidad.
  Retorna dos arreglos x, y con las posiciones (en unidades canónicas) del punto P
  con la aproximación del ACG'''
  x = -a * e * np.cos(M)
  y = 2 * a * e * np.sin(M)
  return (x, y)
````

Vamos a tener en cuenta que M varía entre $[0, 2\pi]$

````python
N = 300
M = np.linspace(0, 2* np.pi, N)
````

Calculamos cómo van a ser las posiciones del planeta con la aproximación del ACG, y hacemos en primer lugar un gráfico 2D para verificar que estemos obteniendo una elipse tal como es esperado.

````python
pos = posicion_ACG(M, a, e)
plt.plot(pos[0], pos[1])
plt.plot(0,0, 'ro', label = 'G')
plt.title('Aproximación del AGC')
plt.grid()
plt.axis('equal')
plt.legend();
````

Observamos que efectivamente la anterior elipse presenta la forma de la figura 1, teniendo en cuenta que tiene como centro al punto G.

Ahora, veamos como es el movimiento del punto G.

En este caso el punto G se mueve siguiendo una circuenferencia de radio a, siguiendo las expresiones:

$$x = a cos(M)$$

$$y = asin(M)$$

````python
def mov_G(M, a):
  '''Recibe un arreglo con los valores de anomalía media en radianes y semieje mayor en unidades
  canónicas.
  Retorna dos arreglos x, y que contienes las posiciones (en unidades canónicas)del punto G con el paso del tiempo'''
  x = a * np.cos(M)
  y = a * np.sin(M)
  return (x,y)
````

Calculamos y graficamos el movimiento del punto G usando la función anterior:

**Comenzamos con el proceso de animación:**

Debemos tener en cuenta que el movimiento del planeta como una elipse con foco en F, es una combinación del movimiento del punto G (alrededor de F) y de P (alrededor de G).

Lo que se espera es que las coordenadas del planeta consignadas en "pos" sean medidas con respecto al punto G. Sin embargo, estos valores están con respecto a un origen (0,0) que vamos a fijar en F, por lo cual, realizaremos un cambio de sistema coordenado para que en cada paso "xp" y "yp" esté medido con respecto a punto G (variable).

El cambio de sistema coordenado consiste de una traslación horizontal y una vertical, en $x_g$ y $y_g$.


es decir:

$$x_{pg} = x_g + x_p$$

$$y_{pg} = y_g + y_p$$

Se debe tener en cuenta que no es suficiente con la traslación realizada, sino que el sistema de referencia en el punto G rota con el tiempo por un ángulo M.

La matriz de rotación asociada para este caso es:

$$\begin{bmatrix}
\cos(M) & -\sin(M) \\
\sin(M) & \cos(M)
\end{bmatrix}$$

Por lo cual:

$$\begin{pmatrix}
x'\\
y'
\end{pmatrix} = \begin{bmatrix}
\cos(M) & -\sin(M) \\
\sin(M) & \cos(M)
\end{bmatrix}  \begin{pmatrix}
x\\
y
\end{pmatrix}
$$

$$x' = x \cos(M) - y \cos(M)$$

$$y' = x\cos(M) + y\cos(M)$$

Además de esto, queremos comparar la predicción de la posición con aquella que se obtiene resolviendo el problema de los dos cuerpos de la ecuación de Kepler.

Tenemos las siguientes ecuaciones:

$$r = \frac{a(1-e^2)}{1 + e\cos (f)}$$

$$x = r \cos(f) - ae$$

$$y = r\sin(f)$$


$$f = 2 \tan^{-1}\left(\sqrt{\frac{1+e}{1-e}}\tan\frac{E}{2}\right)$$

Lo que vamos a hacer es aplicar la función ``` pc.kepler_newton``` que se encuentra en pymcel, para encontrar el valor real que describe el punto P. La idea es posteriormente comparar esto con la aproximación que estamos realizando.

Para ejecutar esta función ingresamos el valor de la anomalía media, la excentricidad, y como punto de prueba ingresamos el valor de la anomalía media nuevamente.

Creamos la siguiente rutina:

````python
def movimiento_p(a, e):
  '''Recibe los valores de semieje mayor en unidades canónicas y la excentricidad.
  Retorna los siguientes arreglos areglos:
  rx: posición x real del punto P en unidades canónicas,
  ry: posición x real del punto P en unidades canónicas,
  r_i: r asociado a cada punto P en unidades canónicas'''
  #cálculo de anomalía excéntrica
  E = np.zeros(N)

  for i in range(N):
    E_val, error, niter = pc.kepler_newton(M[i], e, M[i])
    E[i] = E_val

  #cálculo de anomalía verdadera
  f = np.zeros_like(E)

  for i in range(N):
    f [i] = 2 * np.arctan(np.sqrt((1+e)/(1-e)) * np.tan(E[i] / 2))

  #Cálculo de las coordenadas x, y de la posición de P con la solución a la ecuación de
  rx = np.zeros_like(E)
  ry = np.zeros_like(E)
  ri = np.zeros_like(E)

  for i in range(N):
    r_i = (a* (1-e**2))/(1 + e * np.cos(f[i]))
    ri[i]= r_i
    rx[i] = r_i * np.cos(f[i])
    ry[i] = r_i * np.sin(f[i])

  return rx, ry, ri
````

Ahora, lo que vamos a hacer es obtener las cantidades que necesitamos para realizar la animación:

- Rotación y traslación al punto de referencia G, de la elipse que describe su movimiento sobre la circunferencia centrada en F.

- Ubicación del punto P en el tiempo (coordenadas x, y), como una combinación del movimiento de G y el desplazamiento de la elipse mencionada en el item anterior.


Para cumplir con lo recién mencionado, realizamos la siguiente función:

````python
def punto_y_elipse(a, e):
  '''Recibe los valores de semieje mayor en unidades canónicas y la excentricidad.
  Retorna 4 arreglos:
  p_x: posición x del punto P en el ACG,
  p_y: posición y del punto P en el ACG,
  elipse_movida_x: coordenadas x de la elipse rotada y
  trasladada en cada punto del tiempo,
  elipse_movida_y: coordenadas x de la elipse rotada y
  trasladada en cada punto del tiempo '''
  p_x = np.zeros(N)
  p_y = np.zeros(N)
  elipse_movida_x = np.zeros((N, N))
  elipse_movida_y = np.zeros_like(elipse_movida_x)

  for i in range(N):
    #movimiento de G
    rs_G = mov_G(M, a)
    xg = rs_G[0][i]
    yg = rs_G[1][i]

    #aproximación CCG resp origen
    pos = posicion_ACG(M, a, e)
    elipse_x = pos[0]
    elipse_y = pos[1]

    #aplicamos una rotacion de estas cooordenadas con respecto a un ángulo M
    rotada_x = elipse_x * np.cos(M[i]) - elipse_y * np.sin(M[i])
    rotada_y = elipse_x * np.sin(M[i]) + elipse_y * np.cos(M[i])

    #CCG respecto a G
    elipse_g_x = xg + rotada_x
    elipse_g_y = yg + rotada_y

    elipse_movida_x[i] = elipse_g_x
    elipse_movida_y[i] = elipse_g_y

    #movimiento de P como una suma del movimiento de la elipse y del punto G
    xp = rotada_x[i]
    yp = rotada_y[i]

    p_x[i] = xp + xg
    p_y[i] = yp + yg

  return p_x, p_y, elipse_movida_x, elipse_movida_y
````
