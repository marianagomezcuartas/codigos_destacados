4. El caos en sistemas dinámicos se cuantifica usando una propiedad conocida como los coeficientes de Lyapunov (son varios). El coeficiente de Lyapunov máximo se calcula estudiando la manera como se van separando las trayectorias de dos copias de una partícula, la original y la gemela con el tiempo (ver figura abajo). Es posible probar que la separación entre las trayectorias sigue una ley simple:

   $$
  d(t) = d_0 e^{\gamma(t-t_0)}
  $$

  donde $\gamma$ es el máximo exponente de Lyapunov y $d(t)$ es la distancia entre la posición de la original y la gemela en el tiempo $t$. El valor de $d_0$ es en nuestro caso 0.00001.

  <center><img src="https://drive.google.com/uc?id=1B4tVzSWqJCuZpkwmhsQkhZzcmE8DGid3" align="center" width="550"></center>
  
  El cálculo de este parámetro no es tan simple como parece porque las trayectorias de la original y la gemela pueden diverger significativamente y el valor de $d$ puede llegar a ser inmenso. Para hacerlo más fácil se puede realizar un procedimiento por etapas: se integra la trayectoria por un tiempo pequeño entre $t_0$ y $t_1 = t_0 + \Delta t$; después entre $t_1$ y $t_2 = t_1 + \Delta t$ y así sucesivamente.
  
  Es importante decir que después de que se ha integrado entre $t_0$ y $t_1$ y antes de integrar de $t_1$ a $t_2$ la posición de la partícula gemela debe modificarse así: la nueva gemela debe estar en la línea que une la original y la gemela anterior a una distancia de la original igual a $d_0$.

  En teoría el máximo exponente de Lyapunov es entonces:

  $$
  \gamma=\lim _{n \rightarrow \infty} \sum_{i=1}^n \frac{\ln \left(d_i / d_0\right)}{n \Delta t}
  $$

  Podemos definir un estimado del exponente de Lyapunov después de $n$ saltos como:

  $$
  \gamma_n=\sum_{i=1}^n \frac{\ln \left(d_i / d_0\right)}{n \Delta t}
  $$

Vamos a emplear las ecuaciones 9.3, 9.4, 9.5 del libro *Solar System Dynamics* para encontrar las velocidades en y de las partículas de prueba 1 y 2:

9.3 $$r_1^2 = (x + \mu_2)^2 + y^2$$

9.4 $$r_2^2 = (x - \mu_1)^2 + y^2$$

9.5 $$C_J = n^2 (x^2 + y^2)  + 2 \left(\frac{\mu_1}{r_1} + \frac{\mu_2}{r_2}\right) - \dot x^2 - \dot y^2$$

Teniendo en cuenta que vamos a trabajar en un sistema de unidades canónicas con

- $UM = m_1 + m_2 = M$
- $UL = a$

, entonces $n = 1$

Además, para la aplicación de las ecuaciones 9.3, 9.4 y 9.5 identificamos los siguientes elementos:

- $\mu_1 = 1 - \alpha$
- $\mu_2 = \alpha$
- $x$ y $y$ son las posiciones de la partícula con respecto al origen de coordenadas del sistema de referencia rotante en el tiempo $t=0$
- $r_1$ y $r_2$ son las distancias de la partícula de prueba a las masas $1$ y $2$ respectivamente (diferentes a las partículas de prueba)

Como $y$ para las partículas $1$ y $2$ son cero en ambos casos:

- $r_1 = x + \mu_2$
- $r_2 = x - \mu_1$

Por lo tanto, aplicado lo anterior y despejando $\dot y$ de la ecuación 9.5 tenemos que:

1. $$\dot y = \sqrt{x^2 + 2\left(\frac{\mu_1}{x + \mu_2} + \frac{\mu_2}{x- \mu_1}\right) - \dot x^2 - C_J}$$

Definimos los valores y condiciones iniciales del problema:

````python
alfa = 0.001

#constante de Jacobi
C_j = 3.07

#partícula 1 (p)
r0p = np.array([0.55, 0, 0])

#partícula 2 (s)
r0s = np.array([0.56, 0, 0])
````

Procedemos a hallar el valor de $\dot y$ para las partículas de prueba 1 y 2:


````python
mu_1 = 1 - alfa
mu_2 = alfa

x0p = r0p[0]
x0s = r0s[0]

#disancias de las partículas de prueba a las masas 1 y 2 respectivamente
r1_p = np.sqrt((x0p + mu_2)**2)
r2_p = np.sqrt((x0p - mu_1)**2)
r1_s = np.sqrt((x0s + mu_2)**2)
r2_s = np.sqrt((x0s - mu_1)**2)

#velocidades en y de las partículas de prueba
vel_y_p = np.sqrt(x0p**2 + 2*mu_1/r1_p +  2*mu_2/r2_p - C_j)
vel_y_s = np.sqrt(x0s**2 + 2*mu_1/r1_s +  2*mu_2/r2_s - C_j)
````

Escribimos las condiciones iniciales de velocidad para las partículas de prueba 1 y 2:

````python
#partícula 1 (p)
v0p = np.array([0, vel_y_p, 0])

#partícula 2 (s)
v0s = np.array([0, vel_y_s, 0])
````

Usamos la rutina de pymcel ```pc.constante_jacobi``` para ver si podemos recuperar $C_J = 3.07$, es decir, comprobar que las velocidades que obtuvimos para las partículas 1 y 2 son correctas.

````python
#partícula 1
verif_p = pc.constante_jacobi(alfa, [r0p, r0p], [v0p, v0p])

#partícula 2
verif_s = pc.constante_jacobi(alfa, [r0s, r0s], [v0s, v0s])

verif_p, verif_s
````

Observamos que en ambos casos podemos recuperar la constante de Jacobi de forma correcta, por lo cual procedemos sin ningún problema.

Para la escritura de la función que nos calcule los elementos orbitales, vamos a tener en cuenta el desarrollo realizado en la [minitarea 2](https://colab.research.google.com/drive/1yDJqwj1z47gr6X0WULSNKJlzQF1Qb_Ou?usp=sharing), donde para cada caso en el cual es necesario aplicar un coseno inverso, se tiene en cuenta el intervalo de definición de tal ángulo para establecer condiciones que permitan obtener el ángulo en el cuadrante correcto.

Particularmente se va a encontrar que el ángulo de inclinación en este caso va a ser cero(siempre), y por lo tanto, la linea de los nodos no va a estar definida. Lo que se hace para resolver este problema es ubicarla a la fuerza sobre el eje x, lo cual ocasiona que $\Omega$ y $\omega$ sean cero.

````python
def elem_orbitales(rs, vs, mu=1):
  '''Recibe un vector de posición, velocidad y mu del sistema.
  Retorna a, e, I, W, w, f; donde las distancias están dadas en unidades
  canónicas de distancia y los ángulos son dados en radianes
  '''

  #cálculo de h
  hvec = np.cross(rs, vs)
  h = np.linalg.norm(hvec)

  #determinación de p
  p = h**2 / mu

  #hallar el vector e y su magnitud
  evec = (np.cross(vs, hvec) / mu) - rs / np.linalg.norm(rs)
  e = np.linalg.norm(evec)

  #hallar a
  a = p/ (1-e**2)

  #encontrar la inclinación
  cosI = np.dot(hvec, [0,0,1]) / h
  I = np.arccos(cosI)

  if I > 0:
    nvec = np.cross([0,0,1], hvec)
  else:
    nvec = np.array([1,0,0])

  n = np.linalg.norm(nvec)

  #cálculo de Omega
  Wp = np.arccos(np.dot(nvec, [1,0,0]) / n)    #valor principal
  W = Wp if nvec[1]>=0 else 2*np.pi - Wp

  #cálculo de omega
  wp = np.arccos(np.dot(nvec, evec) / (n * e))
  w  = wp if evec[2] >=0 else 2*np.pi-wp

  #cáclulo de f
  fp =  np.arccos(np.dot(rs, evec) / ((np.linalg.norm(rs)) * e))
  cond = np.dot(rs, vs)
  f = fp if cond >=0 else 2*np.pi-fp

  return a, e, I, W, w, f
````

Tal como se menciona en el libro Murray-Dermott, lo que vamos a hacer es suponer la misma constante de Jacobi y reajustar el valor inicial de velocidad para las partículas:

Ahora, la idea es calcular la excentricidad de las partículas p y s (con sus gemelas) en el sistema de referencia inercial.

Como solo necesitamos la excentricidad, no vamos a usar la rutina ```elem_orbitales```, sino que vamos a crear una rutina que calcule únicamente la excentricidad:

````python
def excentricidad(rs, vs, mu=1):
  '''Recibe un vector de posición, velocidad y mu del sistema.
  Retorna el valor de la excentricidad
  '''
  hvec = np.cross(rs, vs)
  h = np.linalg.norm(hvec)
  p = h**2 / mu
  evec = (np.cross(vs, hvec) / mu) - rs / np.linalg.norm(rs)
  e = np.linalg.norm(evec)
  return e
````

4. El caos en sistemas dinámicos se cuantifica usando una propiedad conocida como los coeficientes de Lyapunov (son varios). El coeficiente de Lyapunov máximo se calcula estudiando la manera como se van separando las trayectorias de dos copias de una partícula, la original y la gemela con el tiempo (ver figura abajo). Es posible probar que la separación entre las trayectorias sigue una ley simple:

   $$
  d(t) = d_0 e^{\gamma(t-t_0)}
  $$

  donde $\gamma$ es el máximo exponente de Lyapunov y $d(t)$ es la distancia entre la posición de la original y la gemela en el tiempo $t$. El valor de $d_0$ es en nuestro caso 0.00001.

  <center><img src="https://drive.google.com/uc?id=1B4tVzSWqJCuZpkwmhsQkhZzcmE8DGid3" align="center" width="550"></center>
  
  El cálculo de este parámetro no es tan simple como parece porque las trayectorias de la original y la gemela pueden diverger significativamente y el valor de $d$ puede llegar a ser inmenso. Para hacerlo más fácil se puede realizar un procedimiento por etapas: se integra la trayectoria por un tiempo pequeño entre $t_0$ y $t_1 = t_0 + \Delta t$; después entre $t_1$ y $t_2 = t_1 + \Delta t$ y así sucesivamente.
  
  Es importante decir que después de que se ha integrado entre $t_0$ y $t_1$ y antes de integrar de $t_1$ a $t_2$ la posición de la partícula gemela debe modificarse así: la nueva gemela debe estar en la línea que une la original y la gemela anterior a una distancia de la original igual a $d_0$.

  En teoría el máximo exponente de Lyapunov es entonces:

  $$
  \gamma=\lim _{n \rightarrow \infty} \sum_{i=1}^n \frac{\ln \left(d_i / d_0\right)}{n \Delta t}
  $$

  Podemos definir un estimado del exponente de Lyapunov después de $n$ saltos como:

  $$
  \gamma_n=\sum_{i=1}^n \frac{\ln \left(d_i / d_0\right)}{n \Delta t}
  $$

  Hacer un gráfico de $\gamma_n$ como función de $t$ para las dos partículas de nuestro problema. Haga un ajuste lineal a la curva y determine la pendiente y el intercepto. ¿Que observa?.

  Se dice que hay caos cuando $\gamma>0$ a largo plazo. ¿Cuál de las dos partículas del problema está en una trayectoria caótica?.

Comenzamos por el definir el valor de $d_0$ y el rango de tiempo a deltas pequeños en el cual vamos a hacer el procedimiento de integración y reescalación que se describe en el procedimiento.

Luego, resolvemos para el arreglo de tiempos recién definido la trayectoria de las partículas p y s.

````python
d0 = 0.00001

d_ts = np.arange(0, 100, 0.02)

#trayectoria de la original p y s
rs_p, vs_p, ris_p, vis_p, r1is_p, r2is_p=  pc.crtbp_solucion(alfa, r0p, v0p, d_ts)
rs_s, vs_s, ris_s, vis_s, r1is_s, r2is_s=  pc.crtbp_solucion(alfa, r0s, v0s, d_ts)
````

Definimos el límite superior y máximo del contador, correspondiente al numero de veces en el cual se va a realizar un reescalamiento.

Para el caso de las partículas gemelas, vamos a hacer el procedimiento tomando de a dos valores de tiempo en el proceso de integración.\
Generamos unos arreglos vacíos, en los cuales se incluyen las condiciones iniciales asociadas al proceso de iteración, y las distancias $d_i$ que van resultando con el tiempo.

Dentro del ciclo for, se realiza el proceso por medio del cual se determina la distancia que separa a su partícula de su gemela en cada tiempo de integración, además de realizar el proceso de reescalación.

````python
#límite del contador
n = len(d_ts)

#condiciones particula de prueba gemela
r0_pg = np.zeros((n, 3))
r0_pg[0] = r0pg
v0_pg = np.zeros((n, 3))
v0_pg[0] = v0pg

r0_sg = np.zeros((n, 3))
r0_sg[0] = r0sg
v0_sg = np.zeros((n, 3))
v0_sg[0] = v0sg

#arreglo de las distancias d_i
ds_ip = np.zeros(n)
ds_ip[0] = d0

ds_is = np.zeros(n)
ds_is[0] = d0

for i in range(n-1):
  ts_ = np.array([d_ts[i], d_ts[i+1]])
  rs_pg, vs_pg, ris_pg, vis_pg, r1is_pg, r2is_pg  =  pc.crtbp_solucion(alfa, r0_pg[i], v0_pg[i], ts_)
  rs_sg, vs_sg, ris_sg, vis_sg, r1is_sg, r2is_sg  =  pc.crtbp_solucion(alfa, r0_sg[i], v0_sg[i], ts_)

  #p
  d_i_vec = rs_pg[1] - rs_p[i+1]
  d_i_norm= np.linalg.norm(d_i_vec)
  ds_ip[i+1] = d_i_norm
  #redefino las condiciones iniciales para la siguiente iteración
  r0_pg[i+1] = (d_i_vec / d_i_norm) * d0 + rs_p[i+1]
  v0_pg[i+1] = vs_pg[1]

  #s
  d_i_vec = rs_sg[1] - rs_s[i+1]
  d_i_norm= np.linalg.norm(d_i_vec)
  ds_is[i+1] = d_i_norm
  #redefino las condiciones iniciales para la siguiente iteración
  r0_sg[i+1] = (d_i_vec / d_i_norm) * d0 + rs_s[i+1]
  v0_sg[i+1] = vs_sg[1]
````

Definimos la función asociada con la fórmula   $\gamma_n=\sum_{i=1}^n \frac{\ln \left(d_i / d_0\right)}{n \Delta t}$.

````python
def gamma_n(ds_i, ts, d_0 = d0):
  '''recibe la separación de las partículas asociado a cada intervalo de tiempo
  (por pares), el tiempo de integración.
  Como argumento opcional, recibe la separación en el tiempo cero de una partícula
  con su gemela.
  Retorna el valor de gamma_n
  '''
  frac = ds_i / d_0
  num = np.log(frac)
  delta_t = d_ts[1] - d_ts[0]
  sum = 0
  for i in range(0, len(ts)):
    sum += num[i-1] / (len(ts) * delta_t)
  return sum
````

Aplicamos la anterior función a las partículas s y p, hallando $\gamma_n$ para n = 1, 2, .....

````python
gammas_p = np.zeros(n)
gammas_s = np.zeros(n)

for i in range(n):
  ts = d_ts[:i+1]
  ds_i_p = ds_ip[:i+1]
  ds_i_s = ds_is[:i+1]
  gammas_p[i] = gamma_n(ds_i_p, ts)
  gammas_s[i] = gamma_n(ds_i_s, ts)
````

Graficamos el resultado obtenido:

````python
plt.subplot(1,2,1)
plt.plot(d_ts, gammas_p)
plt.xlabel('tiempo')
plt.ylabel('$\gamma_n$')
plt.title('partícula p')

plt.subplot(1, 2, 2)
plt.plot(d_ts, gammas_s)
plt.xlabel('tiempo')
plt.ylabel('$\gamma_n$')
plt.title('partícula s')

plt.tight_layout()
````
